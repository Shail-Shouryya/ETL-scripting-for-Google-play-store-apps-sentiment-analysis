```python
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
engine = create_engine("mysql://root:password@localhost/google_play_store")
conn = engine.connect()
```

```python
data = engine.execute("SELECT * FROM ordered_apps_and_reviews")
```

```python
# df = pd.read_sql(query.statement, query.session.bind)
# apps_and_reviews_df = pd.read_sql(data, conn)
apps_and_reviews_df = pd.read_sql("SELECT * FROM ordered_apps_and_reviews", conn)
apps_and_reviews_df
```

```python
# drop any rows with "NaN" values
valid_entries_df = apps_and_reviews_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# resetting index keeps old index and adds new one, but we just want to replace the index
# reset = valid_entries_df.reset_index()

# df.reset_index(drop=True)
reset_index = valid_entries_df.reset_index(drop=True)
reset_index
```

```python
# check to make sure there are no duplicates
reset_index.drop_duplicates
```
