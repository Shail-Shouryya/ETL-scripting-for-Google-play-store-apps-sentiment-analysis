```python
import pandas as pd
```

```python
apps = pd.read_csv('googleplaystore.csv')
apps
```

```python
reviews = pd.read_csv('googleplaystore_user_reviews.csv')
reviews
```

```python
reviews_dropNaN = reviews.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
reviews_dropNaN.reset_index
```

```python
merged_df = pd.merge(apps, reviews_dropNaN, how="right", on='App')
merged_df
```

```python
merged_df = pd.merge(apps, reviews_dropNaN, how="left", on='App')
merged_df
```

```python
merged_df = pd.merge(apps, reviews_dropNaN, how="inner", on='App')
merged_df
```

```python
merged_df.drop_duplicates
```
