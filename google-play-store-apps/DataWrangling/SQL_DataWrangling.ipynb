```python
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
engine = create_engine("mysql://root:password@localhost/google_play_store")
conn = engine.connect()
```

```python
data = engine.execute("SELECT * FROM ordered_apps_and_reviews")
data
```

```python
# df = pd.read_sql(query.statement, query.session.bind)
# apps_and_reviews_df = pd.read_sql(data, conn)
apps_and_reviews_df = pd.read_sql("SELECT * FROM ordered_apps_and_reviews", conn)
apps_and_reviews_df
```

```python
# drop any rows with "NaN" values
valid_entries_df = apps_and_reviews_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# resetting index keeps old index and adds new one, but we just want to replace the index
# reset = valid_entries_df.reset_index()

# df.reset_index(drop=True)
reset_index = valid_entries_df.reset_index(drop=True)
reset_index
```

```python
# check to make sure there are no duplicates
reset_index.drop_duplicates
```

```python
distinct_apps = pd.read_sql("SELECT * FROM distinct_apps", conn)
distinct_apps
```

```python
no_null_reviews = pd.read_sql("SELECT * FROM no_null_reviews", conn)
no_null_reviews.to_csv('no_null_reviews_from_mySQL.csv')
no_null_reviews
```

```python
no_null_or_duplicate_reviews = pd.read_sql("SELECT * FROM no_null_or_duplicate_reviews", conn)
no_null_or_duplicate_reviews.to_csv('no_null_or_duplicate_reviews_fromSQL.csv')
no_null_or_duplicate_reviews
```

```python
distinct_apps_no_null_reviews_merged_df = pd.merge(distinct_apps, no_null_reviews, how='inner', on='App')
distinct_apps_no_null_reviews_merged_df.to_csv('distinct_apps_no_null_reviews_merged_df_fromSQL.csv')
distinct_apps_no_null_reviews_merged_df
```

```python
distinct_apps_no_null_or_duplicate_reviews_merged_df = pd.merge(distinct_apps, no_null_or_duplicate_reviews, how='inner', on='App')
distinct_apps_no_null_or_duplicate_reviews_merged_df.to_csv('distinct_apps_no_null_or_duplicate_reviews_merged_df.csv')
distinct_apps_no_null_or_duplicate_reviews_merged_df
```
